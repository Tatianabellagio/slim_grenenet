```python
## vcf file produced by slim 
vcf = allel.read_vcf('subp1.vcf')
print(len(vcf['variants/POS']))
genotypes = allel.GenotypeArray(vcf['calldata/GT'])
ref_allele_counts = genotypes.count_alleles()[:, 0]
alt_allele_counts = genotypes.count_alleles()[:, 1]
```

```python
slim = pd.read_csv('subp1allele_counts.csv',header=None).rename(columns={0:'pos', 1:'alt'})
slim['pos'] = slim['pos'].astype(int)
## correct for the 0 index of slim
slim.loc[:,'pos'] = slim['pos'] + 1
len(slim)
slim['ref'] = 100 - slim['alt'] 
slim = slim.sort_values(by='pos')[['pos', 'ref', 'alt']].reset_index(drop=True)
```
